### Part 2: Combining and Merging
**Scenario:** You have two DataFrames: `df_students` (containing IDs and Names) and `df_grades` (containing IDs and Subjects).

| df_students | | df_grades |
| :--- | :--- | :--- |
| **ID** | **Name** | **ID** | **Grade** |
| 1 | Alice | 1 | A |
| 2 | Bob | 3 | B |

1.  **Inner Join:** Perform an inner join on `ID`. Which student disappears, and why?
2.  **Left Join:** Perform a left join. What value appears in the `Grade` column for Bob?
3.  **Concatenation:** Use `pd.concat` to stack these two dataframes on top of each other (even if the columns don't match perfectly). Use `ignore_index=True`.


```python
#     **Solution**
# Type your solution here.
df_students = pd.DataFrame({"ID":[1, 2], "Name":["Alice", "Bob"]})
df_grades = pd.DataFrame({"ID":[1,3], "Grade":["A", "B"]})

inner_join = pd.merge(df_students, df_grades, how="inner", on="ID")
print(inner_join)
print("Bob disappears because his ID is not in df_grades\n")

left_join = pd.merge(df_students, df_grades, how="left")
print(left_join)
print("Bob gets a NaN value in the grade column")
```

### Part 3: Reshaping and Pivoting
**Task:** Understanding "Long" vs. "Wide" formats.

1.  **Pivoting:** Given a "long" format dataset (columns: `Date`, `Stock_Ticker`, `Price`), use the `.pivot()` method to make `Date` the index and each `Stock_Ticker` its own column.
2.  **Melting:** Take a "wide" dataframe (columns: `Name`, `Math_Score`, `Science_Score`) and use `pd.melt()` to turn it into a long format with columns `Name`, `Subject`, and `Score`.



```python
#     **Solution**
# Type your solution here.

long_data = pd.DataFrame(columns=["Date", "Stock_Ticker", "Price"])
pivoted = long_data.pivot(index="Date", columns="Stock_Ticker", values="Price")
print(pivoted, "\n")

wide_data = pd.DataFrame(columns=["Name", "Math_Score", "Science_Score"])
melted = pd.melt(wide_data, id_vars=['Name'], var_name='Subject', value_name='Score')
print(melted)
```

Download the `netflix_titles.csv` from Kaggle.

Split the data into two dataframes:

### The Tasks

#### 1. The "Broken" Join (Merging)
Netflix accidentally lost the `show_id` for 5 records in `df_details`. 
* **Task:** Perform a **Left Join** from `df_info` to `df_details`. 
* **Question:** Identify which titles from `df_info` now have `NaN` values for their `rating`. Why is a Left Join better than an Inner Join here if we want to see what data is missing?

#### 2. The "Country" Spread (Reshaping)
The `country` column often contains multiple countries separated by commas (e.g., "United States, India, South Korea").
* **Task:** Use the `.str.split(', ', expand=True)` method to break these into separate columns, then use **`pd.melt()`** to turn this into a "Long" format where every country has its own row for a specific `show_id`.

```python
#     **Solution**
# Type your solution here.

# 1. The Setup & The "Broken" Join
import kagglehub

# Download latest version
path = kagglehub.dataset_download("shivamb/netflix-shows")
#print("Path to dataset files:", path)

df_full = pd.read_csv("/home/codespace/.cache/kagglehub/datasets/shivamb/netflix-shows/versions/5/netflix_titles.csv")

df_info = df_full[["show_id", "type", "title", "country"]]
print("df_info is:\n", df_info, "\n")

df_details = df_full[["show_id", "release_year", "rating", "duration"]]
print("df_details is:\n", df_details, "\n")

left_joined = pd.merge(df_info, df_details, how="left")
print("df_info and df_details left joined is:\n", left_joined)

print("\nAny titles that were not shared between df_info and df_details will have a missing value under the rating column.\n" \
"A left join is better for seeing missing data because an inner join only shows the values shared between data.")
```

```python
#     **Solution**

# Type your solution here.
# 2. The "Country" Spread (Splitting & Melting)

countries = df_info["country"].str.split(',', expand=True)
df_wide = pd.concat([df_info[['show_id', 'title']], countries], axis=1)

melted_countries = pd.melt(df_wide, id_vars=['show_id', 'title'], value_vars=countries.columns, value_name='country', var_name='country_id')

print(melted_countries)
```
